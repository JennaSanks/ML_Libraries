# Pytorch

**What is PyTorch?**

* PyTorch is an open-source deep learning framework developed by Meta (formerly Facebook).
* It is primarily used to build and train neural networks for tasks involving images, text, audio, video, and many other types of data.

* While Scikit-learn is mainly for classical machine learning, PyTorch is designed for deep learning.

**Why is PyTorch used?**

PyTorch helps you:

* Build neural networks
* Train deep learning models
* Use GPUs for faster training
* Perform automatic differentiation (compute gradients automatically)
* Build custom AI architectures
* Research and deploy AI applications

## torch.tensor()

**Why it's used**

Everything in PyTorch is stored as a tensor. Before a model can process data (images, text, numbers, labels), it must be converted into tensors.

```python
import torch

x = torch.tensor([10, 20, 30])
print(x)
```

**Explanation**

* Creates a PyTorch tensor.
* Converts the Python list into a tensor.
* Everything in PyTorch (inputs, outputs, weights, labels) is stored as tensors.

## torch.zeros()

**Why it's used**

Used to initialize tensors with zeros. You'll often use it for creating placeholders, masks, or initializing values before updating them.

```python
x = torch.zeros(2, 3)
print(x)
```

**Explanation**

* Creates a tensor filled with 0.
* 2 = rows, 3 = columns.
* Commonly used to initialize tensors.

# torch.ones()

**Why it's used**

Creates a tensor filled with 1s. Used to initialize weights, masks, or tensors with a constant value, and for testing.

```python
x = torch.ones(2, 3)
print(x)
```

**Explanation**

* Creates a tensor filled with 1.
* Often used for initialization and testing.

## torch.empty()

**Why it's used**

Creates an uninitialized tensor. Used when you'll immediately overwrite the values, making it faster than initializing with zeros or ones.

```python
x = torch.empty(2, 3)
print(x)
```

Explanation

* Creates an uninitialized tensor.
* Values are random because memory is allocated but not filled.
* Faster than zeros() when you'll overwrite the values immediately.

##torch.arange()

**Why it's used**

Creates a sequence of numbers with a fixed step size. Used for ranges, indexing, loops, and generating integer sequences.

```python
x = torch.arange(0, 10, 2)
print(x)
```

Explanation

* Creates equally spaced values.
* Format: start, stop, step.

##torch.linspace()

**Why it's used**

Creates a fixed number of evenly spaced values between two endpoints. Used for sampling continuous values, plotting, and numerical computations.

```python
x = torch.linspace(0, 1, 5)
print(x)
```

Explanation

* Creates a fixed number of evenly spaced values.
* Here it generates 5 numbers between 0 and 1.

##torch.rand()

**Why it's used**

Neural networks usually start with random weights. torch.rand() generates random values between 0 and 1, making it useful for initialization and creating sample data.

```python
x = torch.rand(2, 3)
print(x)
```

**Explanation**

Every time you run it, the values will be different.

##reshape()

```python
x = torch.arange(6)

y = x.reshape(2, 3)

print(y)
```

**Why it's used**

Models expect data in specific shapes. reshape() changes how data is arranged without changing the actual values.



Explanation

**Before:**

    tensor([0, 1, 2, 3, 4, 5])

##torch.randn()

**Why it's used**

Generates random numbers from a normal (Gaussian) distribution with a mean of 0 and a standard deviation of 1.

It is mainly used to initialize weights of neural networks, because values centered around 0 help models train better.

```python
import torch

x = torch.randn(2, 3)
print(x)
```

Output (changes every run)

**Explanation**
* Creates a tensor with 2 rows and 3 columns.
* Values can be positive or negative.
* Numbers are generated from a normal distribution.

##torch.randint()

**Why it's used**

Generates random integers within a given range.

Useful for:

* Creating random class labels.
* Random indices.
* Testing models.

```python
import torch

x = torch.randint(0, 10, (2, 3))
print(x)
```

Output (changes every run)

**Explanation**

* 0 → minimum value (inclusive).
* 10 → maximum value (exclusive).
* (2, 3) → output shape.

The tensor contains random integers from 0 to 9.

##torch.manual_seed()

**Why it's used**

Normally, random functions produce different values every time you run your code.

*torch.manual_seed()* fixes the random number generator so you get the same random values every run.

This makes experiments reproducible.

```python
import torch

torch.manual_seed(42)

x = torch.rand(2, 3)

print(x)
```

Output (will be the same every time)

**Explanation**
* 42 is called the seed.
* You can use any integer (1, 10, 100, 999, etc.).
* The same seed always produces the same sequence of random numbers.

# Tensor Properties

These properties help you understand a tensor's shape, size, type, and where it's stored. You'll use them almost every day when working with PyTorch.

##shape

**Why it's used**

Tells you the dimensions (rows, columns, etc.) of a tensor.

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x.shape)
```

**Explanation**

    2 → rows
    3 → columns

shape is the quickest way to check a tensor's dimensions.

##size()

**Why it's used**

Returns the size (dimensions) of a tensor.

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x.size())
```

**Explanation**

*size()* and *shape* return the same result.

Both are commonly used.

```python
# Example:

print(x.shape)
print(x.size())
```

Both print

    torch.Size([2, 3])

##dtype

**Why it's used**

Tells you what type of data is stored in the tensor.

```python
import torch

x = torch.tensor([1, 2, 3])

print(x.dtype)
```

**Explanation**

Here the tensor stores integers.

```python
# Another example:

x = torch.tensor([1.5, 2.5])

print(x.dtype)
```

**Common data types:**

* torch.int64 → Integers
* torch.float32 → Decimal numbers (most commonly used in deep learning)
* torch.bool → True/False

##device

**Why it's used**

Tells you whether the tensor is stored on the CPU or GPU.

```python
import torch

x = torch.tensor([1, 2, 3])

print(x.device)
```

Output (if using CPU)

**Explanation**

If you're using a GPU, the output may be:

    cuda:0

Knowing the device is important because the model and data must be on the same device.

##ndim

**Why it's used**

Returns the number of dimensions (axes) of a tensor.

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x.ndim)
```

**Explanation**

Examples:

    torch.tensor(5)

Dimensions = 0

    torch.tensor([1, 2, 3])

Dimensions = 1

    torch.tensor([[1, 2],
                  [3, 4]])

Dimensions = 2

    torch.rand(3, 32, 32)

Dimensions = 3

The more axes a tensor has, the higher its dimensions.

##numel()

**Why it's used**

Returns the total number of elements in the tensor.

```python
import torch


x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x.numel())
```

**Explanation**

The tensor has:

* 2 rows
* 3 columns

So,

    2 × 3 = 6 elements

*numel()* simply counts all elements.

#Tensor Data Type Conversion

Sometimes you'll need to change the data type of a tensor. This is very common because many PyTorch functions (especially in deep learning) expect tensors of a specific type.

##float()

**Why it's used**

Converts a tensor to floating-point numbers (float32).

Most neural networks require input data to be float32.

```python
import torch

x = torch.tensor([1, 2, 3])

print(x.dtype)

x = x.float()

print(x.dtype)
```

**Explanation**

Initially, x stores integers.
float() converts every value to a decimal number.
Values become 1.0, 2.0, 3.0 internally.

##int()

**Why it's used**

Converts a tensor to integers.

Useful when decimal values are not needed.

```python
import torch

x = torch.tensor([1.7, 2.8, 3.9])

x = x.int()

print(x)
```

**Explanation**

* Removes the decimal part.
* It **does not round** the number.
* 1.7 → 1
* 2.8 → 2

## long()

**Why it's used**

Converts a tensor to 64-bit integers (int64).

This is the required data type for labels used with many loss functions like CrossEntropyLoss.

```python
import torch

x = torch.tensor([1.2, 2.9])

x = x.long()

print(x)
print(x.dtype)
```

**Explanation**

* Converts values to integers.
* Stores them as int64.

##bool()

**Why it's used**

Converts values into True or False.

Useful for creating masks and filtering data.

```python
import torch

x = torch.tensor([1, 0, 3, 0])

x = x.bool()

print(x)
```

**Explanation**
* 0 becomes False.
* Any non-zero value becomes True.

##type()

**Why it's used**

Converts a tensor to any specific data type.

Less commonly used than .float() or .long(), but useful when you want precise control.

```python
import torch

x = torch.tensor([1, 2, 3])

x = x.type(torch.float32)

print(x.dtype)
```

**Explanation**
* Converts the tensor to the specified type.
* You can replace torch.float32 with other types like torch.int64, torch.bool, etc.

#Tensor Indexing & Slicing

Indexing and slicing let you access or modify specific values in a tensor. This is used constantly while preparing data for machine learning.

##Access a Single Element

**Why it's used**

To get a specific value from a tensor.

```python
import torch

x = torch.tensor([10, 20, 30, 40])

print(x[2])
```

**Explanation**
* Indexing starts from 0.
* x[2] means the 3rd element.

      Index : 0   1   2   3
      Value :10  20  30  40

##Negative Indexing

**Why it's used**

To access elements from the end.

```python
import torch

x = torch.tensor([10, 20, 30, 40])

print(x[-1])
```

**Explanation**

      -1 → Last element
      -2 → Second last
      -3 → Third last

##Slicing

**Why it's used**

To extract multiple elements.

```python
import torch

x = torch.tensor([10, 20, 30, 40, 50])

print(x[1:4])
```

**Explanation

        Start : 1 (included)

        Stop : 4 (excluded)

So it returns

        20 30 40

##Slice from Beginning

```python
import torch

x = torch.tensor([10, 20, 30, 40, 50])

print(x[:3])
```

**Explanation**

Starts from the beginning and stops before index 3.

##Slice till End

```python
import torch

x = torch.tensor([10, 20, 30, 40, 50])

print(x[2:])
```

**Explanation**

Starts at index 2 and goes until the end.

##Step Slicing

**Why it's used**

To skip elements.

```python
import torch

x = torch.tensor([10, 20, 30, 40, 50, 60])

print(x[::2])
```

**Explanation**

    Start → Beginning

    Stop → End

    Step → 2

It picks every second element.

##Indexing 2D Tensors

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x[1, 2])
```

**Explanation**

      Row 1

      Column 2

      [[1,2,3],
      [4,5,6]]
            ↑

##Access a Complete Row

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x[0])
```

**Explanation**

Returns the first row.

##Access a Complete Column

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x[:, 1])
```

**Explanation**

    :  → All rows

    1 → Second column

So it returns

    2

    5

##Modify an Element

**Why it's used**

To update values inside a tensor.

```python
import torch

x = torch.tensor([10, 20, 30])

x[1] = 100

print(x)
```

**Explanation**

The value at index 1 is replaced with 100.

##Modify an Entire Row

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

x[0] = torch.tensor([7, 8, 9])

print(x)
```

**Explanation**

The entire first row is replaced.

##Boolean Indexing

**Why it's used**

To select elements that satisfy a condition.

```python
import torch

x = torch.tensor([10, 20, 30, 40, 50])

print(x[x > 25])
```

**Explanation**

* x > 25 creates a mask:

      [False, False, True, True, True]

* Only elements where the mask is True are returned.

This is heavily used for filtering data.
