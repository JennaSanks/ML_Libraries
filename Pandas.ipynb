**What is Pandas?**

Definition:

Pandas is an open-source Python library used for loading, organizing, cleaning, analyzing, and manipulating structured data.

**It provides two main data structures:**

Series → A one-dimensional labeled array.
DataFrame → A two-dimensional table (like an Excel sheet or SQL table).

Think of Pandas as Excel + SQL + NumPy combined into Python.

**Import it with:**

**1. Import Pandas**

```python
import pandas as pd
```

**Why was Pandas created?**

NumPy is excellent at numerical computation, but it has limitations when working with real-world datasets.

Imagine student data.

Using NumPy:

```python
import numpy as np

students = np.array([
    ["Alice", 20, 85],
    ["Bob", 21, 90],
    ["Charlie", 19, 88]
])

print(students)
```

**Problems:**

*  No column names.
*  Everything becomes strings because NumPy arrays typically have a single data type.
*  Hard to read.
*  Hard to filter by column name.

**Using Pandas:**

```python
import pandas as pd

students = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [20, 21, 19],
    "Marks": [85, 90, 88]
})

print(students)
```

Much cleaner.

**Why?**

pd is the standard alias used by the Python community.

**13. pipe()**

Chain operations.

```python
df.pipe(function_name)
```

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Salary": [50000, 45000, 60000]
}

df = pd.DataFrame(data)

# Create a function
def increase_salary(df):
    df["Salary"] = df["Salary"] * 1.10
    return df

# Pass DataFrame to the function
df = df.pipe(increase_salary)

print(df)
```

Mostly used in advanced projects.

**15. Categorical Data**

**16. Window Functions**

Rolling averages.

```python
import pandas as pd

data = {
    "Day": [1, 2, 3, 4, 5, 6],
    "Sales": [100, 150, 200, 250, 300, 350]
}

df = pd.DataFrame(data)

df["Rolling_Mean"] = df["Sales"].rolling(3).mean()

print(df)
```

Used in finance and time-series.
