**2. sns.lineplot()**

Creates a line graph.

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,4,6,8]

sns.lineplot(x=x, y=y)

plt.show()
```

**4. sns.barplot()**

**Extra**

**1. Figure-level vs Axes-level Functions**

This is the most important extra concept.

*Axes-level*

Draws on a single plot.

```python
import pandas as pd

df = pd.DataFrame({
    "Age": [20, 22, 25, 27, 30],
    "Salary": [30000, 35000, 42000, 48000, 55000]
})
```

```python
sns.scatterplot(data=df, x="Age", y="Salary")
```

*Figure-level*

Creates its own figure and can handle multiple subplots.

```python
sns.relplot(data=df, x="Age", y="Salary")
```

**Examples of figure-level functions:**

* relplot()
* catplot()
* displot()
* jointplot()
* lmplot()

**Examples of axes-level functions:**

* scatterplot()
* lineplot()
* barplot()
* boxplot()
* histplot()
* heatmap()

**9. Exploratory Data Analysis (EDA)**

Most Seaborn usage is during EDA.

A typical workflow:

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "City": ["Pune", "Mumbai", "Pune", "Delhi", "Mumbai"],
    "Marks": [85, 92, 78, 88, 95]
})
```

```python
sns.histplot(df["Marks"])

sns.boxplot(data=df, x="City", y="Marks")

sns.pairplot(df)

sns.heatmap(df.select_dtypes("number").corr(), annot=True)
```

These few plots can reveal a lot about your data before modeling.

**10. Choosing the Right Plot**

| Data Type                                  | Best Plot       |
| ------------------------------------------ | --------------- |
| Trend over time                            | `lineplot()`    |
| Relationship between two numeric variables | `scatterplot()` |
| Category counts                            | `countplot()`   |
| Compare category values                    | `barplot()`     |
| Distribution                               | `histplot()`    |
| Distribution + density                     | `kdeplot()`     |
| Outliers                                   | `boxplot()`     |
| Distribution by category                   | `violinplot()`  |
| Correlation                                | `heatmap()`     |
| Many feature relationships                 | `pairplot()`    |
| Regression trend                           | `regplot()`     |
