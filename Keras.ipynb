# 2. Sequential Model

Creates a neural network layer by layer.


      from tensorflow.keras.models import Sequential

      model = Sequential()

````python
# Example
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create the model
model = Sequential([
    Dense(64, activation="relu", input_shape=(10,)),
    Dense(32, activation="relu"),
    Dense(1, activation="sigmoid")
])

# Display the model architecture
model.summary()
````

(The exact formatting may vary slightly depending on your TensorFlow/Keras version.)

**Explanation**
* Sequential() → Creates the model.
* Dense(64) → First layer with 64 neurons.
* Dense(32) → Second layer with 32 neurons.
* Dense(1) → Output layer with 1 neuron.
* model.summary() → Shows the model structure.

# 3. Dense Layer

Most commonly used layer.

      from tensorflow.keras.layers import Dense

      model.add(Dense(64))

      # Example

      model.add(Dense(64, activation="relu"))

````python
# Example
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(32, input_shape=(4,), activation="relu"))
model.add(Dense(16, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.summary()
````

**Explanation**
* Input has 4 features
* First Dense layer has 32 neurons
* Second Dense layer has 16 neurons
* Output layer has 1 neuron for binary classification

Meaning

* 64 neurons
* ReLU activation

#  4. Input Layer

      from tensorflow.keras import Input

          Input(shape=(10,))

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model = Sequential([
    Input(shape=(4,)),
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.summary()
````

**Explanation**
* Input(shape=(4,)) → Each input has 4 features.
* Dense(8) → Hidden layer with 8 neurons.
* Dense(1) → Output layer with 1 neuron.

#5. Activation Functions

    activation="relu"

    activation="sigmoid"

    activation="softmax"

    activation="tanh"

    activation="linear"

Know when to use each:

| Activation | Used For                   |
| ---------- | -------------------------- |
| ReLU       | Hidden layers              |
| Sigmoid    | Binary classification      |
| Softmax    | Multi-class classification |
| Linear     | Regression                 |
| Tanh       | Sometimes hidden layers    |


````python
# ReLU (Most Common)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model = Sequential([
    Input(shape=(4,)),
    Dense(8, activation="relu")
])

model.summary()
````

**Explanation**
* relu = Rectified Linear Unit
* Converts all negative values to 0
* Keeps positive values unchanged

**When to use?**

Hidden layers (used in almost every deep learning model)

````python
# Sigmoid
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model = Sequential([
    Input(shape=(4,)),
    Dense(1, activation="sigmoid")
])

model.summary()
````

**Explanation**
* Gives an output between 0 and 1
* Perfect for probabilities

**Example:**

    0.92 → 92% chance
    0.08 → 8% chance

**When to use?**

Binary Classification (Yes/No, True/False, Spam/Not Spam)

````python
# Softmax
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model = Sequential([
    Input(shape=(4,)),
    Dense(3, activation="softmax")
])

model.summary()
````

**Explanation**

Returns probabilities for multiple classes, and they add up to 1.

**Example:**

    Cat  : 0.10
    Dog  : 0.75
    Horse: 0.15

Prediction = Dog

**When to use?**

Multi-class Classification

````python
# Tanh
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model = Sequential([
    Input(shape=(4,)),
    Dense(8, activation="tanh")
])

model.summary()
````

**Explanation**

Outputs values between -1 and 1.

**When to use?**

Sometimes in hidden layers (less common than ReLU)

````python
# Linear
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model = Sequential([
    Input(shape=(4,)),
    Dense(1, activation="linear")
])

model.summary()
````

**Explanation**

Returns the output directly without changing it.

**When to use?**

Regression (House Price, Temperature, Stock Price)

# 6. Compile Model

      model.compile(
          optimizer="adam",
          loss="binary_crossentropy",
          metrics=["accuracy"]
      )

````python
# example
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model = Sequential([
    Input(shape=(4,)),
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

print("Model Compiled Successfully!")
````

**Explanation**

    optimizer="adam"

Tells the model how to update its weights to learn better.

    loss="binary_crossentropy"

Measures how wrong the predictions are (used for binary classification).

    metrics=["accuracy"]

Displays the accuracy while training.

**When to use?**

Always call compile() before fit().

###Most Common Combinations

````python
# Binary Classification
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
````

````python
# Multi-class Classification
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
````

````python
# Regression
model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)
````

#7. Optimizers

**What is it?**

An optimizer updates the model's weights to reduce errors and improve predictions.

**Why use it?**

It helps the model learn from its mistakes during training.

Most important:

    Adam()

    SGD()

    RMSprop()

    Adagrad()

Usually

    optimizer="adam"

is enough.

````python
# 1. Adam (Most Recommended)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model = Sequential([
    Input(shape=(4,)),
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
````

**Explanation**
* Fast
* Stable
* Works well in most cases
* When to use?

Almost every deep learning project.

````python
# 2. SGD (Stochastic Gradient Descent)
from tensorflow.keras.optimizers import SGD

model.compile(
    optimizer=SGD(),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
````

**Explanation**
* Learns step by step.
* Slower than Adam but sometimes gives better results.

**When to use?**

When experimenting or learning optimization.

````python
# 3. RMSprop
from tensorflow.keras.optimizers import RMSprop

model.compile(
    optimizer=RMSprop(),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
````

**Explanation**
* Adjusts the learning rate automatically.
* Good for sequence data.

**When to use?**

RNNs and LSTMs.

````python
# 4. Adagrad
from tensorflow.keras.optimizers import Adagrad

model.compile(
    optimizer=Adagrad(),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
````

**Explanation**
* Uses different learning rates for different parameters.
*Good for sparse data.

**When to use?**

Some NLP and recommendation systems.
