**What is Matplotlib?**

Matplotlib is a Python library used to create graphs, charts, and visualizations from data.

Think of it like this:

* NumPy → Stores and performs calculations on data.
* Pandas → Organizes and manipulates data.
* Matplotlib → Turns data into graphs so humans can understand it.

Without Matplotlib, you would only see numbers like:

```python
Marks = [70, 85, 90, 65, 95]
```

With Matplotlib, you can visualize them as a graph, making patterns much easier to understand.

**Why is Matplotlib Used?**

Imagine you have this data:

```python
Marks = [70, 85, 90, 65, 95]
```

Just looking at the numbers doesn't immediately tell you much.

**With Matplotlib:**
* You can see if marks are increasing or decreasing.
* Compare students using a bar chart.
* Find trends over time.
* Detect outliers using a box plot.
* Show proportions using a pie chart.
* Understand data distribution using a histogram.

Humans understand visuals much faster than tables of number

**2. plt.plot()**

Creates a line graph.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,4,6,8]

plt.plot(x, y)

plt.show()
```

**Used for**
* Trends
* Time-series
* Model loss/accuracy graphs

**3. plt.scatter()**

Scatter plot.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [3,5,2,8]

plt.scatter(x, y)

plt.show()
```

* Relationship between two variables

**4. plt.bar()**

Bar chart.

```python
import matplotlib.pyplot as plt

students = ["A","B","C"]
marks = [80,90,75]

plt.bar(students, marks)

plt.show()
```

**Used for**

Comparing categories.

**5. plt.barh()**

```python
plt.barh(students, marks)
plt.show()
```

**6. plt.hist()**

Histogram.

```python
import matplotlib.pyplot as plt

marks = [80,82,85,90,92,75,70,88]

plt.hist(marks)

plt.show()
```

**7. plt.pie()**

```python
import matplotlib.pyplot as plt

labels = ["Python","Java","C++"]
sizes = [50,30,20]

plt.pie(sizes, labels=labels)

plt.show()
```

**27. Object-Oriented API**

Instead of:

```python
plt.plot(x,y)
```

```python
# You can write:

fig, ax = plt.subplots()

ax.plot(x,y)

plt.show()
```

This approach is preferred for larger projects because it's easier to manage multiple plots.

**15. Common Plot Workflow**

This is what you'll do almost every time:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(x, y)

plt.title("Title")

plt.xlabel("X")

plt.ylabel("Y")

plt.grid(True)

plt.legend()

plt.tight_layout()

plt.show()
```
