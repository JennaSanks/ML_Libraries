###tf.constant()

**Definition**

Creates a constant tensor.

A tensor is TensorFlow's basic data structure, similar to a NumPy array.

  Tensor: A multidimensional array used to store and process data in TensorFlow.

**Why is it used?**

Almost everything in TensorFlow (inputs, weights, outputs) is represented as tensors.

**Syntax**

    tf.constant(value)

```python
# Example 1: 1D Tensor
import tensorflow as tf

a = tf.constant([10, 20, 30])

print(a)
```

**Explanation**

* Values → [10, 20, 30]
* Shape → (3,) (3 elements)
* Data type → int32

```python
# Example 2: 2D Tensor
import tensorflow as tf

matrix = tf.constant([
    [1, 2],
    [3, 4]
])

print(matrix)
```

```python
# Example 3: Float Tensor
import tensorflow as tf

a = tf.constant([1.5, 2.7, 3.8])

print(a)
```

```python
# Example 4: String Tensor
import tensorflow as tf

names = tf.constant(["Alice", "Bob"])

print(names)
```

```python
# Example 5: Boolean Tensor
import tensorflow as tf

flags = tf.constant([True, False, True])

print(flags)
```

**Why use tf.constant()?**

Use it when the values should not change.

Examples:

* Input data
* Labels
* Fixed matrices
* Configuration values

###tf.Variable()

**Definition

Creates a mutable tensor.

    Mutable: Can be changed after it is created.

**Why is it used?**

Machine learning models learn by updating their parameters (weights and biases). Those values must be mutable.

**Syntax**

    tf.Variable(value)

```python
# Example 1
import tensorflow as tf

x = tf.Variable(10)

print(x)
```

```python
# Example 2: Updating a Variable
import tensorflow as tf

x = tf.Variable(10)

x.assign(25)

print(x)
```

```python
# Example 3: Adding to a Variable
import tensorflow as tf

x = tf.Variable(5)

x.assign_add(10)

print(x)
```

```python
# Example 4: Subtracting
import tensorflow as tf

x = tf.Variable(20)

x.assign_sub(7)

print(x)
```

**Why use tf.Variable()?**

During training:

    Old Weight → Calculate Gradient → Update Weight

Only tf.Variable allows those updates.

tf.constant() vs tf.Variable()
| Feature           | `tf.constant()` | `tf.Variable()`      |
| ----------------- | --------------- | -------------------- |
| Mutable           | ❌ No            | ✅ Yes                |
| Used for          | Fixed values    | Trainable parameters |
| Can change later? | ❌               | ✅                    |
| Common use        | Input data      | Model weights        |


###tf.convert_to_tensor()

**Definition**

Converts Python objects or NumPy arrays into TensorFlow tens

**Why is it used?**

Many datasets begin as Python lists or NumPy arrays. TensorFlow operations require tensors, so this function converts them.

```python
# Example 1: List → Tensor
import tensorflow as tf

numbers = [1, 2, 3, 4]

tensor = tf.convert_to_tensor(numbers)

print(tensor)
```

```python
# Example 2: NumPy Array → Tensor
import tensorflow as tf
import numpy as np

arr = np.array([10, 20, 30])

tensor = tf.convert_to_tensor(arr)

print(tensor)
```

```python
# Example 3: Specify Data Type
import tensorflow as tf

numbers = [1, 2, 3]

tensor = tf.convert_to_tensor(numbers, dtype=tf.float32)

print(tensor)
```

**When should you use each?**

**tf.constant()**

When you're creating a new tensor from scratch.

    a = tf.constant([1, 2, 3]

**tf.Variable()**

When the value needs to change during training.

    weight = tf.Variable(0.5)

**tf.convert_to_tensor()**

When you already have data in another format.

    import numpy as np

    arr = np.array([1, 2, 3])

    tensor = tf.convert_to_tensor(arr)

###tensor.shape

**Definition**

Returns the shape (dimensions) of a tensor.

**Why is it used?**

Before feeding data into a neural network, you must know its shape. Many TensorFlow errors happen because of incorrect shapes.

**Syntax**

    tensor.shape

```python
# Example 1
import tensorflow as tf

tensor = tf.constant([
    [1,2,3],
    [4,5,6]
])

print(tensor.shape)
```

Meaning:

* 2 rows
* 3 columns

```python
# Example 2
import tensorflow as tf

tensor = tf.constant([1,2,3,4])

print(tensor.shape)
```

*Meaning:*

4 elements in one dimension.

```python
# Example 3 (3D Tensor)
import tensorflow as tf

tensor = tf.constant([
    [[1],[2]],
    [[3],[4]]
])

print(tensor.shape)
```

**Meaning**

    2 blocks

    Each block has

    2 rows

    Each row has

    1 value

###tensor.ndim

**Definition**

Returns the number of dimensions (rank) of the tensor.

    Rank: The number of axes (dimensions) a tensor has.

**Why is it used?**

Deep learning models expect tensors of specific ranks:

* Images → rank 4
* Text sequences → rank 2 or 3
* Single feature vectors → rank 2

**Syntax**

```python
# Example 1
import tensorflow as tf

tensor = tf.constant([1,2,3])

print(tensor.ndim)
```

```python
# Example 2
import tensorflow as tf

tensor = tf.constant([
    [1,2],
    [3,4]
])

print(tensor.ndim)
```

```python
# Example 3
import tensorflow as tf

tensor = tf.constant([
    [[1],[2]],
    [[3],[4]]
])

print(tensor.ndim)
```

**Understanding Dimensions**

Scalar
    tf.constant(5)

    0

**Vector**

    tf.constant([1,2,3])

Rank

    1

**Matrix**

    tf.constant([
        [1,2],
        [3,4]
    ])

Rank

    2

**3D Tensor**

    tf.constant([
        [[1],[2]],
        [[3],[4]]
    ])

Rank

##Tensor Operations

Now that you know what tensors are and how to inspect them, it's time to perform operations on them.

This is equivalent to learning mathematical operations in NumPy.

###tf.add()

**Definition**

Adds two tensors element-wise.

**Element-wise:** The operation is performed on corresponding elements in the tensors.

**Why is it used?**

Addition is used everywhere in deep learning:

* Adding bias to neurons
* Updating values
* Tensor calculations

**Syntax**

    tf.add(x, y)

```python
# Example 1
import tensorflow as tf

a = tf.constant([1,2,3])
b = tf.constant([4,5,6])

result = tf.add(a, b)

print(result)
```

**Explanation**

    1+4=5

    2+5=7

    3+6=9

```python
# Example 2
import tensorflow as tf

a = tf.constant(10)
b = tf.constant(20)

print(tf.add(a,b))
```

###tf.subtract()

**Definition**

Subtracts one tensor from another.

**Syntax**

    tf.subtract(x,y)

```python
# Example
import tensorflow as tf

a = tf.constant([10,20,30])
b = tf.constant([1,2,3])

print(tf.subtract(a,b))
```

###tf.multiply()

**Definition**

Multiplies tensors element-wise.

**Syntax**

    tf.multiply(x,y)

```python
# Example
import tensorflow as tf

a = tf.constant([2,4,6])
b = tf.constant([3,5,7])

print(tf.multiply(a,b))
```

**Explanation**

    2×3=6

    4×5=20

    6×7=42

###tf.divide()

**Definition**

Performs element-wise division.

**Syntax**

    tf.divide(x,y)

```python
# Example
import tensorflow as tf

a = tf.constant([20,30,40])
b = tf.constant([2,3,5])

print(tf.divide(a,b))
```

Notice the result is floating-point.

###tf.matmul()

**Definition**

Performs matrix multiplication.

**Matrix multiplication:**

Combines rows of the first matrix with columns of the second matrix according to linear algebra rules.

**Why is it used?**

This is one of the most important TensorFlow operations.

Every neural network computes something like:

    Output = Input × Weights + Bias

The *Input × Weights* part is matrix multiplication.

**Syntax**

    tf.matmul(a,b)

```python
# Example
import tensorflow as tf

a = tf.constant([
    [1,2],
    [3,4]
])

b = tf.constant([
    [5,6],
    [7,8]
])

print(tf.matmul(a,b))
```

###Matrix Multiplication vs Element-wise Multiplication

```python
# Element-wise

print(tf.multiply(a,b))
```

```python
# Matrix multiplication

print(tf.matmul(a,b))
```

They are completely different operations.

##Reshaping & Manipulating Tensors

###tf.reshape()

**Definition**

Changes the shape of a tensor without changing its data.

**Reshape**:

Rearranges the dimensions while keeping the same values.

**Why is it used?**

Neural networks require inputs of specific shapes.

**Example:**

Original:

    [1,2,3,4,5,6]

    ↓

Reshape

    ↓

    [[1,2,3],
    [4,5,6]]

**Syntax**

    tf.reshape(tensor, new_shape)

```python
# Example 1
import tensorflow as tf

tensor = tf.constant([1,2,3,4,5,6])

new_tensor = tf.reshape(tensor, (2,3))

print(new_tensor)
```

```python
# Example 2
tensor = tf.constant([1,2,3,4,5,6])

print(tf.reshape(tensor,(3,2)))
```

Example 3 (Automatic Size)

Use -1.

```python
tensor = tf.constant([1,2,3,4,5,6])

print(tf.reshape(tensor,(-1,2)))
```

TensorFlow automatically calculates the missing dimension.

**Why -1 is Useful**

Instead of manually counting:

    tf.reshape(x,(128,784))

Use

    tf.reshape(x,(-1,784))

TensorFlow determines the correct number of rows automatically.

Very common in Deep Learning.

###tf.concat()

**Definition**

Joins tensors together.

**Why is it used?**

Common when:

* Combining datasets
* Joining features
* Merging embeddings

**Syntax**

    tf.concat([tensor1,tensor2], axis)

```python
# Example 1
import tensorflow as tf

a = tf.constant([1,2,3])

b = tf.constant([4,5,6])

print(tf.concat([a,b], axis=0))
```

```python
# Example 2 (Rows)
a = tf.constant([
    [1,2],
    [3,4]
])

b = tf.constant([
    [5,6]
])

print(tf.concat([a,b], axis=0))
```

```python
# Example 3 (Columns)
a = tf.constant([
    [1],
    [2]
])

b = tf.constant([
    [10],
    [20]
])

print(tf.concat([a,b], axis=1))
```

**Understanding axis**
    axis=0

    ↓

    Vertical

    ↓

    Add rows

]

    axis=1

    ↓

    Horizontal

    ↓

    Add columns

###tf.stack()

**Definition**

Stacks tensors into a new dimension.

**Difference from concat()**

**concat()** joins existing dimensions.

**stack()** creates a brand-new dimension.

```python
# Example
import tensorflow as tf

a = tf.constant([1,2,3])

b = tf.constant([4,5,6])

print(tf.stack([a,b]))
```

Shape

    (2,3)

```python
# Stack Along Columns
print(tf.stack([a,b], axis=1))
```

Concat vs Stack

Concat

    [1 2 3]

    +

    [4 5 6]

    ↓

    [1 2 3 4 5 6]

Stack

    [1 2 3]

    [4 5 6]

    ↓

    [[1 2 3]
    [4 5 6]]

##Broadcasting

**Definition**

TensorFlow automatically expands smaller tensors during operations.

```python
# Example

import tensorflow as tf

a = tf.constant([1,2,3])

print(a + 10)
```

TensorFlow behaves as if

    10

    ↓

    [10 10 10]

```python
# Another Example

matrix = tf.constant([
    [1,2],
    [3,4]
])

print(matrix + 5)
```

**Why Broadcasting Matters**

Instead of writing

    [10,10,10]

You simply write

    10

TensorFlow does the rest.
