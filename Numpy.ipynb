**1. What is NumPy?**

**Definition:**

NumPy (Numerical Python) is a Python library used for creating and working with arrays and performing fast mathematical operations on them.

Think of NumPy as a supercharged version of Python lists that's designed for numbers.

**You import it like this:**

```python
import numpy as np
```

**Here:**

numpy is the library name.

np is an alias (short nickname).

**2. Why was NumPy created?**

Before NumPy existed, Python only had lists.


```python
numbers = [1, 2, 3, 4, 5]
```


**Lists are great, but they have problems:**

1. Slower for large calculations

2. Can't do element-wise math easily

3. Use more memory

4. Not built for scientific computing

```python
numbers = [1, 2, 3]

print(numbers * 2)
```

**You might expect:**

*[2, 4, 6]*

But Python interprets * 2 for a list as "repeat the list twice."

**Now with NumPy:**

```python
import numpy as np

numbers = np.array([1, 2, 3])

print(numbers * 2)
```


NumPy performs the multiplication on each element.

**19. np.where()**

```python
import numpy as np

a=np.array([2,7,9,4])

print(np.where(a>5))
```

**Meaning:**

Elements greater than 5 are at indices 1 and 2.

You can also replace values conditionally:

```python
result = np.where(a > 5, 100, 0)

print(result)
```

np.where(condition, value_if_true, value_if_false) works like a 'find and replace' based on a condition.

It goes through each item in the array a:

If an item is greater than 5 (the condition is True), it replaces that item with 100 (the value_if_true).

If an item is not greater than 5 (the condition is False), it replaces that item with 0 (the value_if_false).

**1. Broadcasting**

Definition:

Broadcasting is NumPy's ability to perform operations on arrays of different shapes without manually repeating values.

Instead of writing:

[1+5, 2+5, 3+5]

NumPy automatically broadcasts 5 across the array.

```python
# Example
import numpy as np

a = np.array([1, 2, 3])

print(a + 5)
```

```python
import numpy as np

a = np.array([[1],
              [2],
              [3]])

b = np.array([10, 20, 30])

print(a + b)
```

**2. Axis**

This is one of the most confusing concepts.

Consider:

```python
import numpy as np

a = np.array([
    [1,2,3],
    [4,5,6]
])

# Rows ↓
# 1 2 3
# 4 5 6
# Columns →

# axis=0
# Works down the rows (column-wise).
print(np.sum(a, axis=0))

# axis=1
# Works across the columns (row-wise).

print(np.sum(a, axis=1))
```

**Memory Trick**


*   axis=0 → Move down each column.
*   axis=1 → Move across each row.

**4. Copy vs View**

This causes many bugs.

```python
# View
import numpy as np

a=np.array([1,2,3])

b=a

b[0]=100

print(a)
```

Both variables refer to the same array.

```python
# Copy
import numpy as np

a=np.array([1,2,3])

b=a.copy()

b[0]=100

print(a)
print(b)
```

**12. Random Seed**


```python
# Without a seed:
import numpy as np

print(np.random.randint(1,10,5))
```

Every run gives different numbers.

```python
# With a seed:
import numpy as np

np.random.seed(42)

print(np.random.randint(1,10,5))
```
